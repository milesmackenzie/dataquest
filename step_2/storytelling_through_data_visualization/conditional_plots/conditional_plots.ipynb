# Conditional Plots

```python
import pandas as pd
titanic = pd.read_csv("train.csv")
titanic = titanic[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]].dropna()
print (titanic[0:5])
```

```python
# Importing seaborn and using sns.distplot() to generate kernal density plot with histogram bars.

import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(titanic["Age"])
plt.show()
```

```python
# Using .kdeplot to generate line plot with shaded area under the curve.

sns.kdeplot(titanic['Age'], shade=True)
plt.xlabel("Age")
plt.show()
```

```python
# Setting style of plot to white to have coordinate grid hidden and white background color.
# Generating plot, assigning xlabel to "Age" and axis spine with sns.despine().
# Default removes top and right spines, set left and bottom to True to remove.

sns.set_style("white")
sns.kdeplot(titanic["Age"], shade=True)
plt.xlabel("Age")
sns.despine(left=True, bottom=True)
plt.show()
```

```python
# Using the seaborn.FacetGrid object to represent the layout of the 
# plots in the grid and the columns used for subsetting the data.
# Generating three kernal density plots for distribution of ages accross classes.

g = sns.FacetGrid(titanic, col="Pclass", size=6)
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
# Introducing the row parameter to generate a plot of different combinations of unique
# values in the "Survived" and "Pclass" columns.

g = sns.FacetGrid(titanic, col="Survived", row="Pclass")
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
# Generating multiple plots on the same grid by introducing a new parameter, hue.

g = sns.FacetGrid(titanic, col="Survived", row="Pclass", hue="Sex", size=3)
g.map(sns.kdeplot, "Age", shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
# Adding legend.

g = sns.FacetGrid(titanic, col="Survived", row="Pclass", hue="Sex", size=3)
g.map(sns.kdeplot, "Age", shade=True).add_legend()
sns.despine(left=True, bottom=True)
plt.show()
```

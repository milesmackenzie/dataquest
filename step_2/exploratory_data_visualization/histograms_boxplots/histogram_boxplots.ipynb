```python
import pandas as pd
import matplotlib.pyplot as plt
reviews = pd.read_csv('fandango_scores.csv')
```

```python
# Selecting columns to work with and setting it to new DataFrame

cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
norm_reviews = reviews[cols]
```

```python
# Counting frequency of each rating to a 10th of a point for
# Fandango and IMDB

freq_counts_fan = norm_reviews['Fandango_Ratingvalue'].value_counts()
fandango_distribution = freq_counts_fan.sort_index()

freq_counts_imdb = norm_reviews['IMDB_norm'].value_counts()
imdb_distribution = freq_counts_imdb.sort_index()
print(fandango_distribution[0:3])
print(imdb_distribution[0:3])
```

```python
# Generating a histgram of the data with .hist(), default 10 bins.

fig, ax = plt.subplots()

ax.hist(norm_reviews['Fandango_Ratingvalue'], range=(0,5))
plt.show()
```

```python
# Generating four subplot histograms of the ratings for
# Fandango_Ratingvalue, RT_user_norm, Metacritic_user_nom and
# IMDB_norm

fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.hist(norm_reviews['Fandango_Ratingvalue'], 20, range=(0,5))
ax1.set_title('Distribution of Fandango Ratings')

ax2.hist(norm_reviews['RT_user_norm'], 20, range=(0,5))
ax2.set_title('Distribution of Rotten Tomatoes Ratings')

ax3.hist(norm_reviews['Metacritic_user_nom'], 20, range=(0,5))
ax3.set_title('Distribution of Metacritic Ratings')

ax4.hist(norm_reviews['IMDB_norm'], 20, range=(0,5))
ax4.set_title('Distribution of IMDB Ratings')

ax1.set_ylim(0,50)
ax2.set_ylim(0,50)
ax3.set_ylim(0,50)
ax4.set_ylim(0,50)
plt.show()
```

```python
# Generating box plot using .boxplot()

fig, ax = plt.subplots()

ax.boxplot(norm_reviews['RT_user_norm'])
ax.set_ylim(0,5)
ax.set_xticklabels(['Rotten Tomatoes'])
plt.show()
```

```python
# Generating four box plots for each column in num_cols.
# Selecting multiple columns to pass through .boxplot().
# Rotating the xtick labels to make them readable.

num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']

fig, ax = plt.subplots()

ax.boxplot(norm_reviews[num_cols].values)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_ylim(0,5)
plt.show()
```
